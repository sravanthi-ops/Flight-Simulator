```python
#How to count cuisines in each row by splitting on ","?

# defining a function to count the number of cuisines for each restaurant
def cuisine_counter(cuisines):
    if pd.notna(cuisines):
        # spliting the cuisine string by ","
        cuisine_list = cuisines.split(', ')
        # counting the number of cuisines
        return len(cuisine_list)
    else:
        return 0  # you are handling the missing values by returning 0

# applying the cuisine_counter function to create a new column
df['CuisineCount'] = df['Cuisines'].apply(cuisine_counter)

df.head()
```

```python
#How to select relevant features for a continuous target variable using statistical tests and visualizations?

# import necessary library for data analysis and visualization

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Assuming you have loaded your DataFrame (named df)

# converting 'Yes' and 'No' to 1 and 0 for categorical columns
categorical_columns = ['Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu']
df[categorical_columns] = df[categorical_columns].replace({'Yes': 1, 'No': 0})

# Step 1: Create a scatter plot to visualize the relationship between 'Votes' and 'Rating'/Continuous vs. Continuous

sns.scatterplot(x='Votes', y='Rating', data=df)
plt.title('Scatter Plot of Votes vs. Rating')
plt.figure(figsize=(2, 1))
plt.show()

# Step 1: Create another scatter plot, this time for 'Average Cost for two' and 'Rating'/Continuous vs. Continuous

sns.scatterplot(x='Average Cost for two', y='Rating', data=df)
plt.title('Scatter Plot of Average Cost for two vs. Rating')
plt.figure(figsize=(2, 1))
plt.show()


# Step 2: Create box plots and perform ANOVA for categorical vs. continuous variables

categorical_vars = [ 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu']

for var in categorical_vars:
    plt.figure(figsize=(2, 1))
    
    sns.boxplot(x=var, y='Rating', data=df)
    plt.title(f'Box Plot of {var} vs. Rating')
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.figure(figsize=(2, 1))
    plt.show()

    # Performing ANOVA test only if there are more than one category
    categories = df[var].unique()
    valid_categories = [cat for cat in categories if len(df[df[var] == cat]) > 1]
    if len(valid_categories) > 1:
        anova_results = f_oneway(*[df['Rating'][df[var] == category] for category in valid_categories])
        print(f'ANOVA for {var}: F-statistic = {anova_results.statistic}, p-value = {anova_results.pvalue}')


# displaying the first few rows of the modified DataFrame
print(df.head())
```

```python
#Are there any outliers in the data? What about missing values? If any of either, treat them.

# import Pandas library for data manipulation and analysis
import pandas as pd

# Defining the columns of interest for outlier detection
columns_of_interest = ['Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu',
                       'Price range', 'Votes', 'Average Cost for two', 'Rating']

# Defining a multiplier to adjust the outlier detection threshold

outlier_multiplier = 1.5  # Adjusting the value as needed

# Creating a dictionary to store the counts of outliers for each column

outlier_counts = {}  # Dictionary to store outlier counts by column

# Loop through the columns of interest to detect and handle outliers


for column in columns_of_interest:
    # Treating outliers for the current column
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - outlier_multiplier * IQR
    upper_bound = Q3 + outlier_multiplier * IQR

    # Calculating and store the number of outliers for the current column
    outliers = (df[column] < lower_bound) | (df[column] > upper_bound)
    outlier_counts[column] = outliers.sum()

    # Removing rows with outliers for the current column
    df = df[~outliers]

# Displaying outlier counts for specific columns
print("Outlier Counts by Column:")
for column in columns_of_interest:
    print(f"{column}: {outlier_counts[column]} outliers")

# Displaying the first few rows of the modified DataFrame (outliers removed)
print("\nModified DataFrame (Outliers Removed):")
print(df)
```

```python
#On the partitioned data, build the best regression model. Show the accuracy numbers. Which type
#of regression did you choose and why? 

# Import necessary libraries for data manipulation, modeling, and evaluation
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Load your dataset into a DataFrame (replace 'your_dataset.csv' with the actual file)

# Define your features (X) and the target variable (y) for the regression model

X = df[['Price range', 'Votes', 'Average Cost for two','CuisineCount','Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu']]
y = df['Rating']

# Perform one-hot encoding for categorical variables in the feature set

X = pd.get_dummies(X, columns=['Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=523)

# Create a Linear Regression model for predicting restaurant ratings
model = LinearRegression()

# Train the model using the training data
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Evaluating the model's performance using common regression metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Displaying the evaluation metrics to assess model performance
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2) Score:", r2)
```

```python
#On the partitioned data, build a decision tree. Show the accuracy numbers

# Import the DecisionTreeRegressor module from scikit-learn

from sklearn.tree import DecisionTreeRegressor

# Create a Decision Tree model for regression, setting a random seed for reproducibility
tree_model = DecisionTreeRegressor(random_state=523)

# Train the Decision Tree model using the training data
tree_model.fit(X_train, y_train)

# Make predictions on the testing data using the trained Decision Tree model
y_tree_pred = tree_model.predict(X_test)

# Evaluate the performance of the Decision Tree model using regression metrics
tree_mae = mean_absolute_error(y_test, y_tree_pred)
tree_mse = mean_squared_error(y_test, y_tree_pred)
tree_rmse = np.sqrt(tree_mse)
tree_r2 = r2_score(y_test, y_tree_pred)

# Display the evaluation metrics for the Decision Tree model
print("Decision Tree - Mean Absolute Error (MAE):", tree_mae)
print("Decision Tree - Mean Squared Error (MSE):", tree_mse)
print("Decision Tree - Root Mean Squared Error (RMSE):", tree_rmse)
print("Decision Tree - R-squared (R2) Score:", tree_r2)
```
